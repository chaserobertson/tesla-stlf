```python
import pandas as pd
import numpy as np

h21_24 = pd.read_csv('../data/australian-public-holidays-combined-2021-2024.csv').drop('ID', axis=1)
h21_24
```

```python
h20 = pd.read_csv('../data/australian_public_holidays_2020.csv')
h20
```

```python
h19 = pd.read_csv('../data/australian_public_holidays_2019.csv').drop('Raw Date', axis=1)
h19
```

```python
h17_18 = pd.read_csv('../data/australianpublicholidays-201718.csv')
h17_18
```

```python
juris17_18 = pd.DataFrame.from_records(h17_18['Applicable To'].str.lower().str.split('|'))
h17_18_juris = pd.concat([h17_18, juris17_18], axis=1)
h17_18_clean = h17_18_juris.melt(
    id_vars = h17_18.columns[:-1], 
    value_vars = juris17_18.columns, 
    value_name = 'Jurisdiction')
h17_18_clean = h17_18_clean.drop('variable', axis=1).dropna(subset='Jurisdiction').reset_index(drop=True)
h17_18_clean
```

```python
h17_18_clean.pivot_table(index='Holiday Name', aggfunc=lambda x: len(np.unique(x)))
```

Good enough for me. Let's merge these thangs.

```python
h = pd.concat([h17_18_clean, h19, h20, h21_24])
h
```

```python
h['Date'].astype('str').astype('datetime64[ns]').hist()
```

```python
h.pivot_table(index='Jurisdiction', aggfunc=len)
```

```python
h.to_csv('../data/holidays2017_2024.csv', index=None)
```
